1. Which team scored the most points when playing at home?  


````python
sql_query_1 = '''
SELECT 
    t.team_long_name,
    SUM(m.home_team_goal) AS total_home_goals
FROM 
        Match AS m
    JOIN
        Team AS t
    ON 
        m.home_team_api_id = t.team_api_id
GROUP BY 
    home_team_api_id
ORDER BY 
    total_home_goals DESC;
'''
````
